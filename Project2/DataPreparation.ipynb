```python
import requests
import re
import json
```

```python
base_url = 'http://www.calorizator.ru'
base_products = 'http://www.calorizator.ru/product/all'
base_recipes = 'http://www.calorizator.ru/recipes/all'
```

```python
ingredients = {}
```

```python
# first page
html_ = requests.get(base_products).text
```

```python
tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
for tr in trs:
    tmp = [elem.strip() for elem in re.sub('<.*?>', '', tr).split('\n') if elem.strip()]
    if len(tmp) < 5:
        ingredients[tmp[0]] = 0
    else:
        ingredients[tmp[0]] = tmp[-1]
```

```python
# calories for 100 grams
ingredients
```

```python
i = 1
html_ = requests.get(base_products + '?page=' + str(i)).text
while not re.search('<link rel=\"next\" href=\"/product/all\?page=1\" />', html_):
    tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
    trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
    for tr in trs:
        tmp = [elem.strip() for elem in re.sub('<.*?>', '', tr).split('\n') if elem.strip()]
        if len(tmp) < 5:
            ingredients[tmp[0]] = 0
        else:
            ingredients[tmp[0]] = tmp[-1]
    i += 1
    html_ = requests.get(base_products + '?page=' + str(i)).text
```

```python
recipes = {}
```

```python
# first page
html_ = requests.get(base_recipes).text
```

```python
tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
for tr in trs:
    tmp = re.findall('<a href=\"(.*?)\">(.*?)</a>', tr, re.DOTALL)[1]
    recipes[tmp[1]] = tmp[0]
```

```python
recipes
```

```python
i = 1
html_ = requests.get(base_recipes + '?page=' + str(i)).text
while not re.search('<link rel=\"next\" href=\"/recipes/all\?page=1\" />', html_):
    tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
    trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
    for tr in trs:
        try:
            tmp = re.findall('<a href=\"(.*?)\">(.*?)</a>', tr, re.DOTALL)[1]
            recipes[tmp[1]] = tmp[0]
        except IndexError:
            pass # some products
    i += 1
    html_ = requests.get(base_recipes + '?page=' + str(i)).text
```

```python
true_recipes = {}
count = 1
for key, value in recipes.items():
    print(count)
    html_ = requests.get(base_url + value).text
    tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
    trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
    true_recipes[key] = {}
    for tr in trs:
        tmp = [elem.strip() for elem in re.sub('<.*?>', '', tr).split('\n') if elem.strip()]
        true_recipes[key][tmp[0]] = float(tmp[2])
    count += 1
```

```python
count = 102
for key, value in list(recipes.items())[101:]:
    if not value.startswith('/recipes'):
        count += 1
        continue
    print(count)
    html_ = requests.get(base_url + value).text
    tbody = re.findall('<tbody>.*?</tbody>', html_, re.DOTALL)[0]
    trs = re.findall('<tr.*?>.*?</tr>', tbody, re.DOTALL)
    true_recipes[key] = {}
    for tr in trs:
        tmp = [elem.strip() for elem in re.sub('<.*?>', '', tr).split('\n') if elem.strip()]
        true_recipes[key][tmp[0]] = float(tmp[2])
    count += 1
```
